## 2) Démonstration des 5 graphiques

````python
import random
import bibo_21008

bibo_21008.styled_line(
    x=[1,2,3,4,5],
    y=[4,2,5,3,6],
    title="Styled Line",
    xlabel="X",
    ylabel="Y",
)

xs = list(range(30))
ys = [x + random.uniform(-4, 4) for x in xs]
bibo_21008.styled_scatter(xs, ys, title="Styled Scatter", xlabel="X", ylabel="Y")

bibo_21008.styled_bar(["A","B","C","D"], [12, 7, 15, 9], title="Styled Bar", ylabel="Value")

data = [random.gauss(0, 1) for _ in range(400)]
bibo_21008.styled_hist(data, bins=18, title="Styled Histogram", xlabel="Value")

d1 = [random.gauss(0, 1) for _ in range(120)]
d2 = [random.gauss(1.2, 0.8) for _ in range(120)]
d3 = [random.gauss(-0.6, 1.1) for _ in range(120)]
bibo_21008.styled_box([d1, d2, d3], labels=["G1","G2","G3"], title="Styled Box Plot", ylabel="Value")
````
